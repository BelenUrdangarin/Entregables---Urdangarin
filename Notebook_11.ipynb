#### 1) TRANSFORMACION DE VARIABLES

- La mayor√≠a de los algoritmos de machine learning no admiten `strings` como variables y requieren que las variables sean numericas. Por ese motivo, es necesario convertir las variables categoricas en su representaci√≥n numerica. Para esto hay varias opciones que iremos viendo.

**`LabelEncoder()`**

```python
from sklearn.preprocessing import LabelEncoder # Con Label encoder lo que hacemos es cambiar una categoria que tenia nombre por ejemplo de ciudades en que las ponga 0,1,2,3, ect
```

```python
test_encoder = LabelEncoder() #Ejecutamos la funci√≥n, el label enconder agarra un valor y le pone un numero, independientemente de la cantidad de valores diferentes
```

```python
blackfriday.loc[:, 'City_Category'] = test_encoder.fit_transform(blackfriday['City_Category'])
# Para todas las filas de city category, con el test encoder le da un valor por cada categoria de esa columna
```

```python
blackfriday.head(10) #Comprobamos el cambio
```

**`get_dummies()`**

*get_dummies* convierte una variable categorica en variables ficticias/indicadoras, cada variable se convierte en tantas variables 0/1 como valores diferente tenga.

```python
pd.get_dummies(blackfriday, columns=["Gender"]) #Convierte en valores booleanos donde se encuentre un valor
```

```python
pd.get_dummies(blackfriday["Gender"]) #Donde hay dos falsos es un nan
```

```python
blackfriday[['female', 'male']] = pd.get_dummies(blackfriday["Gender"])
# Lo que hace aca es aplicar el get dummies pero en dos filas nuevas y dejar la columna gender en el documento
```

```python
blackfriday.head()
```

**`OneHotEncoder()`**

El *OneHotEncoder* Transforma los datos categ√≥ricos a un formato que los modelos de aprendizaje autom√°tico pueden comprender y utilizar f√°cilmente . Esta transformaci√≥n permite tratar cada categor√≠a de forma independiente sin implicar relaciones falsas entre ellas.
T√©cnica de preprocesamiento de datos categ√≥ricos que convierte variables categ√≥ricas en una representaci√≥n num√©rica binaria que puede ser utilizada por algoritmos de machine learning.
Convierte cada categor√≠a en una columna distinta con valores 0 o 1, indicando la presencia de esa categor√≠a en cada fila.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
blackfriday['Gender'].unique() #Vemos cuantos valores unicos hay en la categoria Gender
```

```python
gender_encoder = OneHotEncoder()
```

```python
gender_encoder.fit_transform(blackfriday[['Gender']]).toarray()
#Genera un array con ceros y unos en los lugares donde se encuentre una categoria
```

```python
gender_encoder.categories_ #Chequeamos las categorias del gender encoder
```

```python
niveles = gender_encoder.categories_[0].tolist()
#Te permite ver qu√© categor√≠as fueron codificadas y en qu√© orden est√°n representadas en las columnas codificadas.
```

```python
one_hot_gender = pd.DataFrame(gender_encoder.fit_transform(blackfriday[['Gender']]).toarray(), columns=niveles)
```

```python
one_hot_gender

#gender_encoder.fit_transform(...)
#Aplica el OneHotEncoder a la columna 'Gender' del DataFrame blackfriday. Esto convierte cada categor√≠a (por ejemplo, 'Male', 'Female') en una fila de ceros y unos.

#.toarray()
#Convierte la salida (que es una matriz dispersa por defecto) a un array NumPy completo con ceros y unos.

#pd.DataFrame(..., columns=niveles)
#Crea un nuevo DataFrame con ese array, usando las categor√≠as (niveles) como nombres de columnas. As√≠ cada columna representa una categor√≠a con 1 si el registro pertenece a esa categor√≠a, y 0 si no.
```

```python
new_df = pd.concat([blackfriday, one_hot_gender], axis=1)
#concatena la trnasformacion del onehot con el frame viejo
```

```python
new_df.head() #Tenemos lo hecho con label_encoder con lo de onehot y el get_dummies
```

**`binning`**

```python
print(blackfriday.Age.min())
print(blackfriday.Age.max())
#Saca el maximo y el minimo de la categoria Age
```

```python
blackfriday["Age"] = pd.to_numeric(blackfriday["Age"], errors='coerce')

#pd.to_numeric(...): convierte cualquier dato en formato texto a num√©rico.
#errors='coerce': si encuentra un valor que no puede convertir a n√∫mero, lo reemplaza por NaN (valor nulo).
```

```python
bin_age = [10, 17, 70, 80]

labels = ["Adolescente", "Adulto", "Anciano"]

age_categories = pd.cut(blackfriday["Age"], bins=bin_age, labels=labels)

#Crea categorias en la columna age con el comando cut
#Adolescente: edades entre 10 y 17
#Adulto: edades entre 17 y 70
#Anciano: edades entre 70 y 80
```
